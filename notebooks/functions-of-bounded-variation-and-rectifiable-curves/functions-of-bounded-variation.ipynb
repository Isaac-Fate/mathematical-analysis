# Functions of Bounded Variation


```python
from pathlib import Path
import numpy as np
from matplotlib.axes import Axes
import matplotlib.pyplot as plt

plt.style.use("math")
figures_dir = Path("../../tex/figures")
```

A function that is bounded but not of bounded variation.


```python
x = np.linspace(1e-5, 1, 100000)
y = np.cos(1 / x)

ax: Axes
fig, ax = plt.subplots(1, 1)
ax.plot(x, y)
plt.tight_layout()
plt.show()

fig.savefig(
    figures_dir.joinpath("bounded-function-that-is-not-of-bounded-variation.png"),
)
```

```python
x = np.linspace(1e-5, 1, 100000)
y = x * np.cos(np.pi / (2 * x))

ax: Axes
fig, ax = plt.subplots(1, 1)
ax.plot(x, y)
plt.tight_layout()
plt.show()

fig.savefig(
    figures_dir.joinpath("continuous-function-that-is-not-of-bounded-variation.png"),
)
```
