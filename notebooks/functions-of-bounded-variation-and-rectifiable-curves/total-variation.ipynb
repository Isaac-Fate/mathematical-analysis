# Total Variation


```python
from pathlib import Path
import portion
import numpy as np
from matplotlib.axes import Axes
import matplotlib.pyplot as plt

from explot import *

plt.style.use("math")
figures_dir = Path("../../tex/figures")
```

```python
x = np.linspace(0, 2, 100)
intervals = [portion.closedopen(0, 1), portion.closed(1, 2)]

fig, axs = plt.subplots(1, 2)

plot_piecewise_function(
    x=x,
    intervals=intervals,
    functions=[
        lambda x: 1 - x,
        lambda x: -x,
    ],
    ax=axs[0],
    label="$f$",
)

plot_piecewise_function(
    x=x,
    intervals=intervals,
    functions=[
        lambda x: 1 / (1 - x),
        lambda x: -1 / x,
    ],
    ax=axs[1],
    label="$\\frac{1}{f}$",
    color="C1",
)


axs[0].legend()
axs[1].legend()
axs[1].set_ylim([-2, 10])

plt.tight_layout()
plt.show()

fig.savefig(
    figures_dir.joinpath("reciprocal-not-of-bounded-variation.png"),
)
```

```python
x = np.linspace(0, 3, 300)
intervals = [portion.closed(0, 1), portion.openclosed(1, 3)]

fig, ax = plt.subplots(1, 1)

plot_piecewise_function(
    x=x,
    intervals=intervals,
    functions=[
        lambda x: x,
        lambda x: -(x - 1) * (x - 3),
    ],
    ax=ax,
    label="$f$",
)

plt.show()

fig.savefig(
    figures_dir.joinpath("piecewise-function-of-bounded-variation.png"),
)
```
