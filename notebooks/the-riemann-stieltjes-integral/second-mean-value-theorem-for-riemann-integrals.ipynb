# Second Fundamental Theorem of Calculus


```python
from pathlib import Path
import numpy as np
import scipy
from scipy.integrate import quad, romberg

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.axes import Axes

import seaborn as sns

sns.set_theme()
mpl.rc(
    "figure",
    dpi=150,
    # Use tight layout
    autolayout=True,
)
mpl.rc("text", usetex=True)
mpl.rc("font", family="Helvetica")


FIGURES_DIR = Path("../../tex/figures")
```

```python
f = lambda x: np.sin(x) / x
F = lambda x: np.sin(x) / 3

k = 10
x = np.linspace(1e-5, np.pi * k, 1000)

ax: Axes
fig, ax = plt.subplots(1, 1)

y = f(x)
ax.plot(x, y)
ax.fill_between(x, 0, y, where=y >= 0, alpha=0.3)
ax.fill_between(x, y, 0, where=y < 0, alpha=0.3, color="tab:red")


xticks = list(map(lambda i: i * np.pi, np.arange(0, k + 1)))
xtick_labels = [r"$0$", r"$\pi$"] + list(
    map(lambda i: r"$" + str(i) + r"\pi$", np.arange(2, k + 1))
)
ax.set_xticks(xticks, labels=xtick_labels)


plt.show()

# fig.savefig(FIGURES_DIR / "change-of-variables-in-a-riemann-integral-example-1.png")
```
