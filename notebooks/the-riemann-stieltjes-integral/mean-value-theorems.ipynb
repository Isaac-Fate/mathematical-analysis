```python
from pathlib import Path
import numpy as np
from matplotlib.axes import Axes
import matplotlib.pyplot as plt
import dotenv
import os
from fractions import Fraction

from mathplot import *

plt.style.use("math")
figures_dir = Path("../../tex/figures")
```

```python
ax: Axes
fig, ax = plt.subplots(1, 1, figsize=(8, 4))


def thomae(x: float):

    q = Fraction(x).denominator
    return 1 / q


xs = np.linspace(0, 1, 100000)
ys = list(map(thomae, xs))

ax.scatter(xs, ys, s=3)
plt.tight_layout()
plt.show()
```
