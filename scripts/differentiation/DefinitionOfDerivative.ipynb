## One-Sided Derivatives and Infinite Derivatives

```python
epsilon = np.finfo(float).eps
a = 0
b = 0.5
x = np.linspace(max(a, epsilon), b, 5000)
f = lambda x : x**2 * sin(1 / x**2)
df = lambda x : 2 * x * sin(1 / x**2) - (2 / x) * cos(1 / x**2)

fig, ax = plt.subplots(figsize=(16, 8), dpi=300)

ax.plot(
    x, f(x),
    color="C0",
    label="$f(x)$ defined on $[0, 0.5]$"
)

delta = 0.1
f_ext = lambda x : df(b) * (x - b) + f(b)

x = np.linspace(a - delta, a, 100)
ax.plot(
    x, np.zeros_like(x),
    color="C1",
    label="extended line with slope $f^\prime_{+}(0) = 0$"
)

x = np.linspace(b - delta, b, 100)
ax.plot(
    x, f_ext(x),
    color="C2",
    ls="--"
)

x = np.linspace(b, b + delta, 100)
ax.plot(
    x, f_ext(x),
    color="C2",
    label="extended line with slope $f^\prime_{-}(0.5)$"
)

ax.legend()

plt.savefig(
    "../../tex/figures/ma-016.png",
    bbox_inches="tight",
    pad_inches=0
)
plt.show()
```
