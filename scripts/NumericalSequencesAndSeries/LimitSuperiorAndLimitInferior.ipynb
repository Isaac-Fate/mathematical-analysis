```python
import numpy as np
from numpy import exp, pi, sin, cos
import matplotlib.pyplot as plt
```

```python
font = {
    'font.family' : 'serif',
    'font.weight' : 'bold',
    'font.size'   : 16
}
plt.rcParams.update(font)

fig, ax = plt.subplots(figsize=(16, 8), dpi=500)

# plot sequence a_n
a = lambda n : (1 + 10/n) * sin(n * pi / 32)
n = 500
ns = np.arange(1, n+1)
plt.scatter(
    ns, a(ns),
    label="$a_n$",
    s=5
)

# plot sup u_n
N = 1000
Ns = np.arange(1, N+1)
u = [np.max(a(Ns[i:])) for i in range(n)]
plt.scatter(
    ns, u,
    s=1,
    label="$u_n = \sup_{m \geq n} a_m$"
)
plt.hlines(
    1,
    xmin=1, xmax=n,
    colors="C1",
    ls="--",
    label="$\limsup_{n \\to \infty} a_n = 1$"
)

# plot inf l_n
l = [np.min(a(Ns[i:])) for i in range(n)]
plt.scatter(
    ns, l,
    s=1,
    label="$l_n = \inf_{m \geq n} a_m$"
)
plt.hlines(
    -1,
    xmin=1, xmax=n,
    colors="C2",
    ls="--",
    label="$\liminf_{n \\to \infty} a_n = -1$"
)

# plot inf ln
plt.ylim(-1.3, 2.1)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
