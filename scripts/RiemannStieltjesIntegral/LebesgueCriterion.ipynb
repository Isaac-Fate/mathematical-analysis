```python
fig, ax = plt.subplots(1, 1, figsize=(4, 4))

x = np.linspace(0, 1, 100)
y = np.ones_like(x) * 0
ax.plot(
    x, y
)
p = Arc(
    xy=(0.2, 0), 
    width=1, 
    height=2,
    angle=0,
    theta1=-60,
    theta2=60
)
ax.text(0, 0, "$[$", fontsize=32)
ax.text(0.4, 0, "$)$", fontsize=32)
ax.add_artist(p)

ax.set_ylim(-1, 1)
plt.show()
```
