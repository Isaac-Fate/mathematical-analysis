# Change of Variables in Riemann Integrals

```python
import numpy as np
from numpy import exp, log, sin, cos, pi
import scipy
from scipy.integrate import quad, romberg
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.artist import Artist
from matplotlib.legend_handler import HandlerLine2D, HandlerPatch

import sys
sys.path.append("../")
from plt_tools import config_plt, fill_between_3d
config_plt()
```

```python
epsilon = np.finfo(float).eps
a = 0
b = 1
x = np.linspace(a, b, 10000)
f = lambda x : x * sin(1 / x)
phi = lambda x : x * (x - 1) * (x - 2)
d_phi = lambda x : 3 * x**2 - 6 * x + 2
integrand = lambda x : f(phi(x)) * d_phi(x)

fig, ax = plt.subplots(figsize=(16, 8), dpi=300)

ax.plot(
    x, integrand(x),
    color="C0",
    label="$f[\\phi(x)] \\phi^\\prime(x)$"
)

integrand_pos = lambda x : np.clip(integrand(x), a_min=0, a_max=np.Inf)
ax.fill_between(
    x,
    0, integrand_pos(x),
    alpha=0.3,
    color="C1",
    label="positive area"
)

integrand_neg = lambda x : np.clip(integrand(x), a_min=-np.Inf, a_max=0)
ax.fill_between(
    x,
    0, integrand_neg(x),
    alpha=0.3,
    color="C2",
    label="negative area"
)


ax.legend()

plt.savefig(
    "../../tex/figures/ma-017.png",
    bbox_inches="tight",
    pad_inches=0
)
plt.show()
```
